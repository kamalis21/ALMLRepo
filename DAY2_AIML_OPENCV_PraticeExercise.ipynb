```python
import cv2

# Load the image
image = cv2.imread("C:\\Users\\hp\\Pictures\\pexels-jack-hawley-57905.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Unable to load the image.")
else:
    # Define the kernel size for Gaussian blur (must be an odd number)
    kernel_size = (5, 5)

    # Apply Gaussian blur to the image
    gaussian_blur_image = cv2.GaussianBlur(image, kernel_size, 0)

    # Define the parameters for the bilateral filter
    d = 9  # Diameter of each pixel neighborhood
    sigma_color = 75  # Filter sigma in the color space
    sigma_space = 75  # Filter sigma in the coordinate space

    # Apply bilateral filter to preserve edges
    filtered_image = cv2.bilateralFilter(gaussian_blur_image, d, sigma_color, sigma_space)

    # Define the alpha (blending) value for overlay
    alpha = 0.7

    # Blend the original image and the filtered image
    overlaid_image = cv2.addWeighted(image, 1 - alpha, filtered_image, alpha, 0)

    # Display the original, blurred, filtered, and overlaid images
    cv2.imshow('Original Image', image)
    cv2.imshow('Gaussian Blurred Image', gaussian_blur_image)
    cv2.imshow('Filtered Image', filtered_image)
    cv2.imshow('Overlaid Image', overlaid_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread("C:\\Users\\hp\\Pictures\\pexels-jack-hawley-57905.jpg")

cv2.imshow('Original Grayscale Image', image)

# Define kernel for operations
kernel = np.ones((5, 5), np.uint8)

# Erosion: Erodes away the boundaries of foreground objects
erosion_image = cv2.erode(image, kernel, iterations=1)

cv2.imshow('Erosion Image', erosion_image)

# Dilation: Expands the boundaries of foreground objects
dilation_image = cv2.dilate(image, kernel, iterations=1)

cv2.imshow('Dilation Image', dilation_image)

# Opening: Erosion followed by dilation, useful for noise reduction
opening_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

cv2.imshow('Opening Image', opening_image)

# Closing: Dilation followed by erosion, useful for closing small holes in objects
closing_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

cv2.imshow('Closing Image', closing_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
