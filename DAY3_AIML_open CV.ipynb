```python
import cv2
import numpy as np

# Load the image
image = cv2.imread("C:\\Users\\hp\\Pictures\\pexels-pixabay-276374.jpg")

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray, 50, 150)

# Find lines using Hough Transform on the edges
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, 150)

# Draw lines on the original image
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Concatenate the original image and the Canny edges horizontally
result = np.hstack((image, cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)))

# Display the result
cv2.imshow("Original Image and Canny Edges", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Read the image in grayscale
image = cv2.imread("C:\\Users\\hp\\Pictures\\pexels-pixabay-276374.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Use Canny edge detection
edges = cv2.Canny(blurred, 50, 150)

# Use HoughLinesP to detect lines
lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=50, minLineLength=50, maxLineGap=30)

# Create a black and white mask based on the maxLineGap condition
mask = np.zeros_like(image)

# Draw lines on the mask
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(mask, (x1, y1), (x2, y2), 255, 2)  # Draw white lines on the black mask

# Set pixels in the original image based on the mask
result = cv2.bitwise_and(image, mask)

# Display the result
cv2.imshow('Result with maxLineGap', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Create an image with half pixels having a value of 120 and the other half with a value of 128
image_height = 720
image_width = 1280
image = np.zeros((image_height, image_width), dtype=np.uint8)

# Set the first half of the pixels to 120
image[:, :image_width//2] = 120

# Set the second half of the pixels to 128
image[:, image_width//2:] = 128

# Set a threshold value between 120 and 128
threshold_value = (120 + 128) // 2

# Apply binary thresholding
_, binary_threshold = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Display the result
cv2.imshow("Original Image", image)
cv2.imshow("Binary Threshold", binary_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Assuming you have an image with 720 pixels
total_pixels = 720

# Specify the percentages and values
percentage_158 = 0.05  # 5%
percentage_120 = 1 - percentage_158  # 95%
value_158 = 158
value_120 = 120

# Calculate the threshold value based on the weighted average
threshold_value = int(percentage_158 * value_158 + percentage_120 * value_120)

# Create an image with the specified percentages and values
image = np.zeros((1, total_pixels), dtype=np.uint8)
image[:, :int(percentage_158 * total_pixels)] = value_158
image[:, int(percentage_158 * total_pixels):] = value_120

# Apply binary thresholding
_, binary_threshold = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Display the result
cv2.imshow("Original Image", image.reshape(1, -1))
cv2.imshow("Binary Threshold", binary_threshold.reshape(1, -1))
cv2.waitKey(0)
cv2.destroyAllWindows()

print("Calculated Threshold Value:", threshold_value)
```
