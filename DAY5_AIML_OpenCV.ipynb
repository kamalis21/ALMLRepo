# To split the frames from a video and save them as individual images

```python
import cv2
import os

# Function to split video into frames
def split_frames(video_path, output_folder):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get video properties
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    for frame_number in range(frame_count):
        # Read the frame
        ret, frame = cap.read()
        if not ret:
            break

        # Save the frame as an image
        frame_filename = os.path.join(output_folder, f"frame_{frame_number:04d}.png")
        cv2.imwrite(frame_filename, frame)

    # Release the video capture object
    cap.release()

    print(f"Frames extracted: {frame_count}")
    print(f"Frames per second (FPS): {fps}")
    print(f"Frames saved in: {output_folder}")
video_path = 'output_video.mp4'


output_folder = 'output_frames'


split_frames(video_path, output_folder)
```

```python
import cv2
import numpy as np

# Open the input video file
input_video_path = "C:\\Users\\hp\\output_video.mp4"  # Replace with the path to your input video file
cap = cv2.VideoCapture(input_video_path)

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the duration to apply the blur (in seconds)
blur_duration = 5

# Create a VideoWriter object to save the output video
output_video_path = 'output_blurred_video.mp4'  # Replace with the desired output video path
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Initialize variables
frame_count = 0

# Loop through the video frames
while cap.isOpened():
    ret, frame = cap.read()

    # Check if the frame was successfully read
    if not ret:
        break

    # Apply blur for the specified duration
    if frame_count < blur_duration * fps:
        frame = cv2.GaussianBlur(frame, (15, 15), 0)

    # Save the modified frame to the output video
    out.write(frame)

    frame_count += 1

# Continue playing the video normally after the blur duration
while cap.isOpened():
    ret, frame = cap.read()

    # Check if the frame was successfully read
    if not ret:
        break

    # Save the unmodified frame to the output video
    out.write(frame)

# Release the video capture and writer objects
cap.release()
out.release()

print(f"Blurred video saved to {output_video_path}")

# Optionally, you can add code to close any open windows if using cv2.imshow
cv2.destroyAllWindows()
```

```python
import cv2

def trim_video(input_path, output_path, start_time_seconds, end_time_seconds):
    # Open the input video file
    vid = cv2.VideoCapture(input_path)

    # Check if the video opened successfully
    if not vid.isOpened():
        print("Error: Could not open video.")
        return

    # Get video properties
    fps = int(vid.get(cv2.CAP_PROP_FPS))
    width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Create a VideoWriter object to save the trimmed video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # Calculate the starting and ending frames based on the given times
    start_frame = int(start_time_seconds * fps)
    end_frame = int(end_time_seconds * fps)

    # Set the video capture to the starting frame
    vid.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    # Read and write frames from the starting frame to the ending frame
    while vid.isOpened() and start_frame <= end_frame:
        ret, frame = vid.read()

        if not ret:
            break

        # Write the frame to the output video
        out.write(frame)

        start_frame += 1

    # Release the video capture and writer objects
    vid.release()
    out.release()

    print(f"Trimmed video saved to {output_path}")

# Get input video path from the user
input_path = input("Enter the path to the input video file: ")

# Get output video path from the user
output_path = input("Enter the path to save the trimmed video: ")

# Get start and end times from the user in seconds
start_time_seconds = float(input("Enter the start time (in seconds) for trimming: "))
end_time_seconds = float(input("Enter the end time (in seconds) for trimming: "))

# Call the function to trim the video
trim_video(input_path, output_path, start_time_seconds, end_time_seconds)
```
